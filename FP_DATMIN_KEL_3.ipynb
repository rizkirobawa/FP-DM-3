```python
import re

# Fungsi untuk mendeteksi emoji
def contains_emoji(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticon
        "\U0001F300-\U0001F5FF"  # simbol & gambar
        "\U0001F680-\U0001F6FF"  # transportasi & simbol
        "\U0001F700-\U0001F77F"  # simbol tambahan
        "\U0001F780-\U0001F7FF"  # tambahan baru
        "\U0001F800-\U0001F8FF"  # tambahan baru
        "\U0001F900-\U0001F9FF"  # tambahan baru
        "\U0001FA00-\U0001FA6F"  # tambahan baru
        "\U0001FA70-\U0001FAFF"  # tambahan baru
        "\U00002702-\U000027B0"  # simbol lain
        "\U000024C2-\U0001F251"  # simbol beragam
        "]+",
        flags=re.UNICODE,
    )
    return bool(emoji_pattern.search(text))

# Memproses kolom teks untuk mendeteksi emoji
columns_to_check = ['full_text']  # Kolom yang akan diperiksa
for column in columns_to_check:
    data = data[~data[column].astype(str).apply(contains_emoji)]

# Menyimpan hasil tanpa baris dengan emoji
# cleaned_file_path = "./cleaned_ppn12.csv"
# data.to_csv(cleaned_file_path, index=False, encoding='utf-8')

# print(f"File tanpa baris dengan emoji berhasil disimpan: {cleaned_file_path}")
```

```python
!pip install nltk
pd.set_option('max_colwidth', 180)

# CLEANING DATA
def clean_text(text):
    text = text.lower()
    text = re.sub(r'@[^\s]+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'#[^\s]+', '', text)
    text = re.sub(r'\b[a-zA-Z]\b', '', text)

    return text

clean = lambda x: clean_text(x)

dfx = pd.DataFrame(data.full_text.apply(clean))
dfx
```

```python
pd.set_option('max_colwidth', 180)
def Punctuation(string):
    punctuations = '''!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~'''

    for x in string.lower():
        if x in punctuations:
            string = string.replace(x, "")

    return string

cleanPunc = lambda x: Punctuation(x)

dfx = pd.DataFrame(dfx.full_text.apply(cleanPunc))
dfx
```

```python
# STOPWORD REMOVAL
def get_stopword(stopwordsfile):
    stopwords = []
    file_stopwords = open(stopwordsfile, 'r')
    row = file_stopwords.readline()
    while row:
        word = row.strip()
        stopwords.append(word)
        row = file_stopwords.readline()
    file_stopwords.close()
    return stopwords

stop_words_indo = get_stopword('stopwords.txt')
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import pandas as pd

nltk.download('punkt_tab')

def stopwords(text):
  tokens = word_tokenize(text)
  filtered = []

  for w in tokens:
    if w not in stop_words_indo:
      filtered.append(w)

  result = ' '.join(filtered)
  return result

st = lambda x: stopwords(x)

dfx = pd.DataFrame(dfx.full_text.apply(st))
dfx
```

```python
!pip install Sastrawi

# Load positive and negative lexicons
def load_lexicons(positive_path, negative_path):
    positive_lexicon = pd.read_csv(positive_path, sep='\t', names=['word', 'weight'])
    negative_lexicon = pd.read_csv(negative_path, sep='\t', names=['word', 'weight'])

    # Convert weights to numeric (if not already)
    positive_lexicon['weight'] = pd.to_numeric(positive_lexicon['weight'], errors='coerce').fillna(0)
    negative_lexicon['weight'] = pd.to_numeric(negative_lexicon['weight'], errors='coerce').fillna(0)

    # Convert to dictionary: word -> weight
    positive_dict = dict(zip(positive_lexicon['word'], positive_lexicon['weight']))
    negative_dict = dict(zip(negative_lexicon['word'], negative_lexicon['weight']))

    return positive_dict, negative_dict

def analyze_sentiment(text, positive_dict, negative_dict):
    words = text.split()
    score = 0
    for word in words:
        pos_weight = positive_dict.get(word, 0)
        neg_weight = negative_dict.get(word, 0)
        print(f"Word: {word}, Positive Weight: {pos_weight}, Negative Weight: {neg_weight}")  # Debugging
        score += pos_weight
        score += neg_weight

    # Determine sentiment label
    if score > 0:
        return 'positive', score
    elif score < 0:
        return 'negative', score
    else:
        return 'neutral', score

# Load dataset
dataset_path = 'cleanedDataPPN12.csv'
data = pd.read_csv(dataset_path)

# Load lexicons
positive_path = 'positive.tsv'
negative_path = 'negative.tsv'
positive_dict, negative_dict = load_lexicons(positive_path, negative_path)

# Apply sentiment analysis
results = data['full_text'].apply(lambda x: analyze_sentiment(str(x), positive_dict, negative_dict))
data['sentiment_label'] = results.apply(lambda x: x[0])  # positive, negative, or neutral
data['sentiment_score'] = results.apply(lambda x: x[1])  # numerical score
```

```python
# LOAD LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.style.use('ggplot')
sns.set_style('whitegrid')

import csv
import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
import pickle

# Vectorized
vectorizer = TfidfVectorizer(
    use_idf=True,             # Gunakan Inverse Document Frequency
    strip_accents='ascii',    # Hapus aksen dari karakter
    min_df=1,                 # Kata muncul minimal di 1 dokumen
    max_df=0.95               # Hapus kata yang terlalu umum (>95% dokumen)
)

# 2. Data Splitting
y = data.sentiment_label       # Label target (Positive/Negative)
x = data.full_text           # Data input (teks)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=33
)

# 3. Vectorization
x_train_vect = vectorizer.fit_transform(x_train)
x_test_vect = vectorizer.transform(x_test)

# Naive Bayes

# 4. Model Training
mnb = MultinomialNB()
mnb.fit(x_train_vect, y_train)

# 5. Prediction
y_pred = mnb.predict(x_test_vect)

# 6. Evaluation
accuracy = accuracy_score(y_test, y_pred)
print("Naive Bayes Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Simpan model dan vectorizer ke file pickle
with open("classifierMNB.pkl", "wb") as model_file:
    pickle.dump(mnb, model_file)

with open("vectorizerMNB.pkl", "wb") as vectorizer_file:
    pickle.dump(vectorizer, vectorizer_file)
```

```python
cm = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Positive'],
            yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Naive Bayes')
plt.show()
```

```python
# Get predicted probabilities for the positive class
y_pred_proba = mnb.predict_proba(x_test_vect)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label='positive')  # Specify pos_label
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--', lw=2, color='red',label='Random Guess')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer

# Load model dan vectorizer
with open("classifierMNB.pkl", "rb") as model_file:
    classifier = pickle.load(model_file)

with open("vectorizerMNB.pkl", "rb") as vectorizer_file:
    vectorizer = pickle.load(vectorizer_file)

def predict_sentiment(text):
    # Transform input text menggunakan vectorizer
    text_vectorized = vectorizer.transform([text])
    # Prediksi menggunakan model Random Forest
    prediction = classifier.predict(text_vectorized)
    return prediction[0]

def main():
    st.title("Prediksi Sentimen Terhadap Kenaikan PPN 12%")
    st.subheader("Input teks untuk mendapatkan prediksi sentimen: positive, negative, atau neutral.")

    # Input text dari user
    user_input = st.text_input("Masukkan teks:", placeholder="Contoh: Kenaikan PPN ini terlalu tinggi!")

    if st.button("Prediksi"):
        if user_input.strip() == "":
            st.error("Teks tidak boleh kosong!")
        else:
            sentiment = predict_sentiment(user_input)
            st.success(f"Hasil prediksi: **{sentiment}**")

if __name__ == "__main__":
    main()
```
